```python
import ctypes
```

```python
class Meralist:
    def __init__(self):
        self.size = 1 # default capacity
        self.n = 0 # count of actual elements
    #create a  c type array with size = self.size
        self.A = self.__make_array(self.size)
    
    def __getitem__(self, index):
        if index < 0 or index >= self.n:
            raise IndexError('Index out of bounds')
        return self.A[index]
    
    def clear(self):
        self.size = 1
        self.n = 0
       
    def find(self, item):
        for i in range(self.n):
            if self.A[i] == item:
                return i
        return "not found in the list"
    
    def pop(self):
        if self.n== 0:
            raise IndexError('Pop from empty array')  
        print(self.A[self.n-1])      
        self.n -= 1
        
        
    def insert(self, index, item):
        if index < 0 or index > self.n:
            raise IndexError('Index out of bounds')
        if self.n == self.size:
            self.__resize(2*self.size) # double the size of the array if capacity is reached
        for i in range(self.n, index, -1):
            self.A[i] = self.A[i-1]
        self.A[index] = item
        self.n += 1
        
    def remove(self, item):
        pos = self.find(item)
        if type(pos) == int:
            self.__delitem__(pos)
        else:
            raise ValueError('Item not found in the list')
     
        
    def __len__(self):
        return self.n
    def append(self,item):
        if self.n == self.size:
            #we will resize the array to double the size of the current array
            self.__resize(2*self.size) # double the size of the array if capacity is reached
        self.A[self.n] = item
        self.n += 1

    def __str__(self):
        #return the string representation of the array
        result = ''
        for i in range(self.n):
            result += str(self.A[i]) + ','  
        return '[' + result[:-1] + ']'
    
    def min(self):
        if self.n == 0:
            raise ValueError('Min from empty array')
        min_value = self.A[0]
        
            
        for i in range(1, self.n):
            if type(self.A[i]) == int:
                if self.A[i] < min_value:
                    min_value = self.A[i]
        return min_value
    
    def __delitem__(self, index):
        if index < 0 or index >= self.n:
            raise IndexError('Index out of bounds')
        for i in range(index, self.n-1):
            self.A[i] = self.A[i+1]
        self.n -= 1
    
    def merge_unique(self, list1):
        for i in range(len(list1)):
            if list1[i] not in self:
                self.append(list1[i])
    
        
    
    def __resize(self, new_capacity):
        #create a new array with the new capacity
        B = self.__make_array(new_capacity)
        #copy the old elements to the new array
        for i in range(self.n):
            B[i] = self.A[i]
        #set the new array as the current array
        self.A = B
        self.size = new_capacity
        
    def __make_array(self, capacity):
        #creates a c type array(static,referential)
        return (capacity * ctypes.py_object)()
```
